```python
import matplotlib.pyplot as plt
import numpy as np
import math

import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.arange(-25, 25, 0.4)
y = (3*(x**4) + 1)/( x**3 )

plt.plot(x, y, '-r', label='y=xxxx')
plt.title('Graph of y=xxxx')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
x = np.arange(-15, 14, 0.5)
y = (math.e**(x+1))/( x+1 )

plt.plot(x, y, '-r', label='y=2x+1')
plt.title('Graph of y=2x+1')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
xAxis = np.linspace(-20, 20, 100)
i1 = np.where(xAxis>0)
i2 = np.where(xAxis<0)
plt.plot(xAxis[i1],(3*(xAxis[i1]**4) + 1)/( xAxis[i1]**3 ))
plt.plot(xAxis[i2],(3*(xAxis[i2]**4) + 1)/( xAxis[i2]**3 ))
plt.title('Graph of y=xxxx')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
xAxis = np.linspace(-5, 5, 500)
i1 = np.where(xAxis>-1)
i2 = np.where(xAxis<-1)
plt.plot(xAxis[i1],(math.e**(xAxis[i1]+1))/( xAxis[i1]+1 ))
plt.plot(xAxis[i2],(math.e**(xAxis[i2]+1))/( xAxis[i2]+1 ))
plt.title('Graph of y=xxxx')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
